# Principi teorici

Il simbolo di π fu introdotto per la prima volta nei testi matematici nel 1706 grazie al matematico gallese W.Jones
(1675-1749) nella sua opera A new Introduction to the Mathematics. Jones indicava con tale simbolo il rapporto
fra la lunghezza di una circonferenza e il suo diametro (c = πd). Nel 1761,J.H. Lambert(1728-1777) dimostrò che
πè un numero irrazionale e quindi non è esprimibile come rapporto di due numeri interi.π è anche un numero
trascendente,come dimostrato nel 1882 da F. von Lindemann (1852-1939),ciò significa che non esistono polinomi
a valori razionali di cui xè una radice. Il numero non può essere espresso tramite un numero finito di numeri interi
o frazioni; è invece composto da una serie infinita di numeri decimali non periodica. La sequenza dei suoi primi
dieci numeri decimali è la seguente: 3.1415926535. Nonostante non esista termine alla sequenza di cifre
decimali, il loro calcolo è una sfida tecnologica e informatica che alimenta conoscenze e miglioramenti nei settori
della matematica applicata. Nel 2021 si è stabilito un nuovo record di cifre calcolate di π.Dopo circa 108 giorni di
calcolo, l'uso di due microprocessori di ultima generazione e di TB di memoria, si è riusciti a calcolare 62800
miliardi di cifre dopo la virgola.

Da un punto di vista storico (senza nessuna pretesa di completezza) le prime approssimazioni di π risalgono ai babilonesi (tavoletta di Susa,1600 a.C.), che approssimavano il numero con la frazione 25/8=3.125.L'accuratezza di questa stima di π è di circa lo 0.5%, valore assolutamente accettabile sotto molti aspetti pratici.
Un contributo nel calcolo di πdi estremo interesse è a opera di Archimede(250 a.C.).Il contributo del siracusano non è solo importante perché raggiunse una approssimazione di π con uno scarto dello 0.03%,ma perché adotta una strategia nota come metodo di esaustione,largamente adoperato nelle dimostrazioni scientifiche. In breve Archimede inscrisse e circoscrisse la circonferenza con poligoni di ugual numero di lati; l'approssimazione del calcolo della circonferenza migliora se si aumenta il numero di lati dei poligoni scelti (partendo d l'esagono, Archimede arrivò a usare poligoni di 96 lati!). Il valore fornito da Archimede fu π=3,141eccezionale accuratezza.

La definizione di π come rapporto fra le lurghezze di ci conferenza e diametro di un cerchio (c = πd) è già sfruttabile operativamente.

La differenza fra il valore di π ottenuto (da approssimazioni matematiche o misure sperimentali) e il valore di
π attualmente noto, alla stessa cifra decimale e assunto come esatto, definisce il livello di accuratezza della
misura. Più la misura è accurata, minore sarà questa differenza.

## Formule legate a π
$C=Circonferenza$

$D=Diametro$

$$ $$
$$π=\dfrac{Circonferenza}{Diametro}=\dfrac{C}{D}$$
$$ $$

$\Delta C = Incertezza \ di \ C$ data dalla sensibilità del mentro

$\Delta D = Incertezza \ di \ D$ data dalla sensibilità del calibro

$E_{R}(C)= errore \ relativo \ di \ C = \dfrac{\Delta C}{C}$

$E_{R}(D)= errore \ relativo \ di \ D = \dfrac{\Delta D}{D}$

L'errore relativo di una grandezza o un mnumero calcolato tramite divisione, si ottienne somando gli errori relativi di ciò che è al numratore e di ciò che è al denominatore, perciò:

$$ $$

$$E_{R}(\pi) = (E_{R}(D)+E_{R}(C))$$

$$ $$

e di conseguenza, dato che $ E_{R}(\pi) = \dfrac{\Delta \pi}{\pi}$, allora l'incertezza su π è:


$$ $$
$$ \Delta \pi = E_{R}(\pi) * \pi .$$

Genera un grafico a dispersione con barre di errore e retta di regressione lineare utilizzando i dati del file "misura_circonferenza.csv". L'asse x deve rappresentare il "Diametro (D) [cm]" con le relative barre di errore "∆D [cm]", e l'asse y deve rappresentare la "Circonferenza (C) [CM]" con le relative barre di errore "∆C [cm]". ## Pulizia e conversione dei dati

### Subtask:
Convertire le colonne "Diametro (D) [cm]", "∆D [cm]", "Circonferenza (C) [cm]" e "∆C [cm]" in tipi numerici, gestendo eventuali errori di formato.


**Reasoning**:
I need to convert the specified columns to numeric types, handling potential format errors. The provided steps require selecting the columns, replacing commas with periods, converting to float, and removing rows with NaN values. I will combine these steps into a single code block for efficiency.



```python
cols_to_convert = ['Diametro (D) [cm]', '∆D [cm]', 'Circonferenza (C) [cm]', '∆C [cm]']
for col in cols_to_convert:
    df[col] = df[col].astype(str).str.replace(',', '.')
    df[col] = pd.to_numeric(df[col], errors='coerce')

df.dropna(subset=cols_to_convert, inplace=True)

display(df.head())
```

## Creazione del grafico

### Subtask:
Generare un grafico a dispersione con "Diametro (D) [cm]" sull'asse x e "Circonferenza (C) [cm]" sull'asse y.


**Reasoning**:
Create a scatter plot of 'Circonferenza (C) [cm]' against 'Diametro (D) [cm]' with appropriate labels and a title.



```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(df['Diametro (D) [cm]'], df['Circonferenza (C) [cm]'])
plt.xlabel('Diametro (D) [cm]')
plt.ylabel('Circonferenza (C) [cm]')
plt.title('Circonferenza vs Diametro')
plt.grid(True)
plt.show()
```

## Aggiunta delle barre di errore

### Subtask:
Includere le barre di errore sull'asse x utilizzando i valori di "∆D [cm]" e sull'asse y utilizzando i valori di "∆C [cm]".


**Reasoning**:
Modify the plot to include error bars for both x and y axes using the specified columns.



```python
plt.figure(figsize=(8, 6))
plt.errorbar(df['Diametro (D) [cm]'], df['Circonferenza (C) [cm]'],
             xerr=df['∆D [cm]'], yerr=df['∆C [cm]'], fmt='o', capsize=3)
plt.xlabel('Diametro (D) [cm]')
plt.ylabel('Circonferenza (C) [cm]')
plt.title('Circonferenza vs Diametro con Barre di Errore')
plt.grid(True)
plt.show()
```

## Regressione lineare

### Subtask:
Eseguire una regressione lineare per trovare la relazione tra il diametro e la circonferenza.


**Reasoning**:
Perform linear regression on the diameter and circumference data to find the relationship.



```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

X = df['Diametro (D) [cm]'].values.reshape(-1, 1)
y = df['Circonferenza (C) [cm]'].values

# Perform linear regression without intercept (forcing the line through the origin)
model = LinearRegression(fit_intercept=False)
model.fit(X, y)

slope = model.coef_[0]
intercept = model.intercept_ # This will be 0 if fit_intercept=False

print(f'Coefficiente angolare (slope): {slope}')
print(f'Intercetta (intercept): {intercept}')

# Calculate R-squared
y_true = y
y_pred = model.predict(X)
r_squared = r2_score(y_true, y_pred)
print(f'Coefficiente di determinazione (R²): {r_squared}')
```

## Plot della retta di regressione

### Subtask:
Aggiungere la retta di regressione al grafico.


**Reasoning**:
Generate predicted circumference values using the linear regression equation and add the regression line to the existing scatter plot with error bars.



```python
# Generate predicted values
predicted_circumference = slope * df['Diametro (D) [cm]'] + intercept

# Plotting
plt.figure(figsize=(8, 6))
plt.errorbar(df['Diametro (D) [cm]'], df['Circonferenza (C) [cm]'],
             xerr=df['∆D [cm]'], yerr=df['∆C [cm]'], fmt='o', capsize=3, label='Data with Error Bars')

# Add the regression line
# Explicitly show the equation with C and D and the slope as pi
plt.plot(df['Diametro (D) [cm]'], predicted_circumference, color='red', label=f'Regression Line: C = {slope:.4f} * D')

plt.xlabel('Diametro (D) [cm]')
plt.ylabel('Circonferenza (C) [cm]')
plt.title('Circonferenza vs Diametro con Barre di Errore e Regressione Lineare')
plt.grid(True)
plt.legend()
plt.show()
```

## Visualizzazione dell'equazione

### Subtask:
Visualizzare l'equazione della retta di regressione sul grafico o in un output separato.


**Reasoning**:
Stampare l'equazione della retta di regressione lineare formattata con il coefficiente angolare e l'intercetta calcolati in precedenza.



```python
# Explicitly show the equation with C and D and the slope as pi
print(f'Equazione della retta di regressione: C = {slope:.4f} * D')
print(f'Coefficiente di determinazione (R²): {r_squared:.4f}')
```
